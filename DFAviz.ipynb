```python
from automata.fa.dfa import DFA

from visual_automata.fa.dfa import VisualDFA
```

```python
room = DFA(
    states={'s0', 's1', 's2'},
    input_symbols={"*x**", "*E**", "***x", "***S", "x***", "N***"},
    transitions={
        's0': {'*x**': 's0', '*E**': 's1' },
        's1': {'***x': 's1', '***S': 's2' },
        's2': {'x***': 's2', 'N***': 's1' },
    },
    initial_state='s0',
    final_states={'s0'},
    allow_partial=True
)
```

```python
visual_room = VisualDFA(room)
visual_room.show_diagram()
```

```python
maze = DFA(
    states={'s0', 's1', 's2', 's3'},
    input_symbols={"*x**", "*E**", "***x", "***S", "x***", "N***", "**x*","**W*"},
    transitions={
        's0': {'x***': 's1', 'N***': 's2' },
        's1': {'*x**': 's3', '*E**': 's0' },
        's2': {'**x*': 's0', '**W*': 's3' },
        's3': {'***x': 's2', '***S': 's1' },
    },
    initial_state='s0',
    final_states={'s0'},
    allow_partial=True
)
```

```python
visual_maze = VisualDFA(maze)
visual_maze.show_diagram()
```
